```python
df = pd.read_csv('transactions.csv')
```

```python
pairs = pd.read_pickle('new_pairs_all_rests')
df_pairs_frequencies = pd.DataFrame(pd.Series(pairs.index).apply(lambda x: x.split(' ')))
```

```python
df_pairs_frequencies['frequency'] = pairs.values
```

```python
df_pairs_frequencies.columns = ['pair','frequency']
```

```python
cohort = df_pairs_frequencies.loc[(df_pairs_frequencies['frequency']>= 10)].copy() 
```

```python
all_people_cohort = []

for p in cohort['pair'].values:
    if p[0] not in all_people_cohort:
        all_people_cohort.append(p[0])
    if p[1] not in all_people_cohort:
        all_people_cohort.append(p[1])
```

```python
len(all_people_cohort)
```

```python
from pylab import rcParams
```

```python
list_consecutive = []

name_prev = 'x'
for name,group in df.groupby(['SHOP','u_day','CONFIGURATIONID']):
    if name[0]!=name_prev:
        print(name[0])
        name_prev = name[0]

    #same shop,same line,same day
    i = 0
    
    for n,transaction in group.groupby('TRANSACTIONID'):
        if i == 0:
            i +=1
            transaction_prev = transaction
            continue
        
        if transaction_prev['PERSONID'].unique()[0]!=0 and transaction['PERSONID'].unique()[0]!=0  and transaction['PERSONID'].unique()[0]!= transaction_prev['PERSONID'].unique()[0]:
            
            #check if both people in the cohort
            if str(transaction_prev['PERSONID'].unique()[0]) in all_people_cohort and str(transaction['PERSONID'].unique()[0]) in all_people_cohort:
            
                time_first = transaction_prev[['year', 'month', 'day', 'hour','minute','second']].apply(lambda s : datetime.datetime(*s),axis = 1).iloc[0]
                time_second = transaction[['year', 'month', 'day', 'hour','minute','second']].apply(lambda s : datetime.datetime(*s),axis = 1).iloc[0]

                if (time_second-time_first).seconds < 5*60:
                    entry = {}
                    entry['person1'] = transaction_prev['PERSONID'].unique()[0]
                    
                    entry['shop'] = name[0]
                    entry['person2'] = transaction['PERSONID'].unique()[0]
                    entry['time1'] = time_first
                    entry['time2'] = time_second
                    
                    entry['products1'] = transaction_prev['lowname'].values
                    entry['categories1'] = transaction_prev['category'].values
                    entry['widecategories1'] = transaction_prev['wide_cat'].values
                    
                    entry['products2'] = transaction['lowname'].values
                    entry['categories2'] = transaction['category'].values
                    entry['widecategories2'] = transaction['wide_cat'].values

                    list_consecutive.append(entry)

        transaction_prev = transaction
```

```python
len(list_consecutive)
```

```python
df_isolated_pairs = pd.DataFrame(list_consecutive)
```

```python
def sorting_ids(x):
    y = sorted([int(x[0]),int(x[1])])
    return str(y[0])+" "+str(y[1])
```
